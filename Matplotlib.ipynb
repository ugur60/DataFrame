```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(1,6)
```

```python
x
```

```python
y = np.arange(2,12,2)
```

```python
y
```

```python
plt.plot(x,y,"red")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"red")
plt.subplot(2,2,2)
plt.plot(y,x,"blue")
plt.subplot(2,2,3)
plt.plot(x,y+2,"green")
plt.subplot(2,2,4)
plt.plot(x,y*4,"pink")
plt.show()
```

```python
fig=plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.2,0.3])
axes1.plot(y,x)
axes1.set_xlabel("Outer X")
axes1.set_ylabel("İnner Y")
axes1.set_title("İnner Graph")
axes2.plot(x,y)
axes2.set_xlabel("Outer Y")
axes2.set_ylabel("İnner X")
axes2.set_title("Outer Graph")
```

```python
x
```

```python
y
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)
for ax in axes:
    ax.plot(x,y)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)
axes[0].plot(x,y)
axes[0].set_title("İlk Grafik")
axes[1].plot(y*2,x**2)
axes[1].set_title("İkinci Grafik")
plt.tight_layout()
```

```python
fig= plt.figure(figsize=(1,1))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color="yellow")
plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,8))
axes[0].plot(x,y,color="red")
axes[0].set_title("ilk")
axes[1].plot(x,y**0.5,color="yellow")
axes[1].set_title("son")
plt.show()
plt.tight_layout()
```

```python
fig= plt.figure(figsize=(6,4))
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**0.5,color="#7FFF00",linewidth=3,linestyle ="-.",marker=10,label=("X'in Karekökü"))
axes.plot(x,x**2,color="#DC143C",linewidth=3,linestyle ="-.",marker="o",label=("X'in Karesi"))
axes.plot(x,x**3,color="#006400",linewidth=3,linestyle ="-.",marker="o",label=("X'in Küpü"))
axes.legend()
axes.set_xlim(0,10)
axes.set_ylim(0,50)
plt.show()
```
