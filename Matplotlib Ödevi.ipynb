```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(51)
y = np.arange(0,101,2)
z = x ** 3
```

```python
fig= plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_xlim(0,50)
axes.set_ylim(0,100)

plt.plot(x,y,"red",linewidth=3,linestyle ="-",marker=10)
axes= fig.add_axes([0,0,1,1])

axes.set_xlabel("X Values")
axes.set_ylabel("Y Values")
axes.set_title("X & Y Graph")

plt.show()
```
